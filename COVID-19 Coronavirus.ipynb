```python
# rename column iso 3166-1 alpha 3 CODE to Iso_Code
df.rename(columns={'ISO 3166-1 alpha-3 CODE':'Iso_Code'},inplace=True)
df.head(4)
```

```python
#top 5 countries which have more number of covid cases
top_5_countries=df.nlargest(n=5,columns="Total Cases")
top_5_countries
```

```python
#plot bar graph for top 5 countries
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 4), dpi=75)
sns.barplot(data=top_5_countries, x="Country", y="Total Cases")
plt.title("Top 5 Countries with Highest Total Cases")
plt.xlabel("Country")
plt.ylabel("Total Cases")

plt.show()
```

```python
#top 5 countries which have more number of covid Deaths
top_5_countries=df.nlargest(n=5,columns="Total Deaths")
top_5_countries
```

```python
#plot bar graph for top 5 countries which have more number of covid Deaths
plt.figure(figsize=(10, 4), dpi=75)
sns.barplot(data=top_5_countries, x="Country", y="Total Deaths")
plt.title("Top 5 Countries with Highest Total Deaths")
plt.xlabel("Country")
plt.ylabel("Total Deaths")

plt.show()
```

```python
#5 countries which have least number of covid Death
least_5_countries=df.nsmallest(n=5,columns="Total Deaths")
least_5_countries
```

```python
#plot bar graph for top 5 countries which have least number of covid Deaths
plt.figure(figsize=(10, 4), dpi=75)
sns.barplot(data=least_5_countries, x="Country", y="Total Deaths")
plt.title("Least 5 Countries with Highest Total Deaths")
plt.xlabel("Country")
plt.ylabel("Total Deaths")

plt.show()
```

```python
#Death percentage
#top 5 countries which have more number of covid  Death percentage 
top_5_death_per=df.nlargest(n=5,columns="Death percentage")
top_5_death_per
```

```python
#plot bar graph for top 5 countries which have more number of covid  Death percentage 
plt.figure(figsize=(10, 4))
sns.barplot(data=top_5_death_per, x="Country", y="Death percentage")
plt.title("Top 5 Countries with Highest Total Deaths")
plt.xlabel("Country")
plt.ylabel("Total Deaths")

plt.show()
```

```python
#continent wise cases (hint use group by)
case_continent=df.groupby('Continent').sum()
case_continent
```

```python
#draw bar plot for continent wise case
case_continent = case_continent.reset_index()
plt.figure(figsize=(10,8), dpi=122)
sns.barplot(data=case_continent, x="Continent", y="Total Cases")
plt.title("Top 5 Continent  with Highest Total Case")
plt.xlabel("Continent")
plt.ylabel("Total Case")

plt.show()
```

```python
#find relationship between Population and total cases.Comment  on outliers
plt.figure(figsize=(15, 10))
plt.xlabel('Population', fontsize=15)
plt.ylabel('Total Cases', fontsize=15)
plt.title('Scatter Plot of Population vs Total Cases', fontsize=15)

sns.scatterplot(data=df, x='Population', y='Total Cases', color='black')

plt.show()
```

```python
#find relationship between Population and total cases.Comment  on outliers
plt.figure(figsize=(15, 10))
plt.xlabel('Total Cases', fontsize=15)
plt.ylabel('Total Deaths', fontsize=15)
plt.title('Scatter Plot of Population vs Total Deaths', fontsize=15)

sns.scatterplot(data=df, x='Total Cases', y='Total Deaths', color='black')

plt.show()
```

```python
# Spearman rank correlation heatmap.Comment on co-relation.

df.corr(method="spearman")# selecting the method as a sperman
plt.figure(figsize=(20,8))# setting the figuresize

heatmap = sns.heatmap(df.corr(method = 'spearman').round(3), vmin=-1, 
vmax=1, annot=True)# annot = True means writting the data value in each cell.

font2= {'family':'serif','color':'green','size':20}
plt.title("Spearman Rank Correlation",font2)
plt.show()# displaying heatmap
```
