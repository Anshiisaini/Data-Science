# Handaling Outliers
 

```python
 #printing quantile range
Q1=dff.quantile(0.25).round(3)   #taking lower quantile into Q2
Q3=dff.quantile(0.75).round(3)   #taking upper quantile to Q3
IQR=Q3-Q1  #calculating inter quantile range
print(IQR)  #printing quantile range
```

```python
dfout=dff[~((dff<(Q1-1.5*IQR))|(dff>(Q3+1.5*IQR))).any(axis=1)] # removing outliers from whole data set
dfout.shape
```

# Model building

In model building we have to prepare the data for ML algorithm. Also we have to prepare the data in a sach a way that it gives better efficiancy for stability of data set we can drop the highly co related features in the data set 

```python
x= dfout.drop(['Salary'], axis=1) #dropping highly co-related features
```

```python
x #new dataset
```

```python
y=dfout['Salary']
```

```python
#split x and y into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=0)
x_train
```

```python
y_train
```

```python
x_test
```

```python
#data normalization with sklearn
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
#fit scaler on training data
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

# Machine Learning techniques

```python
from sklearn.tree import DecisionTreeRegressor #importing DecisionTreeRegressor from sklearn.tree library
#It seems like you've created a DecisionTreeRegressor instance with a maximum depth of 3. If you have further questions or if you want to perform specific operations using this regressor, feel free to provide more context or ask your question!
dt = DecisionTreeRegressor(max_depth=3) 
```

```python
dt.fit(x_train, y_train) #fitting x_train, y_tarin
```

```python
#by using dt.predict function on x_test and calling it on y_predict to predict a  value
y_pred=dt.predict(x_test)
#the code here provided is making use of the predict function to generate predictions based on the input data x_test
print(y_pred)#checking the predicted values
```

```python
#importing ,metrics from sklearn library
from sklearn import metrics 
#metrics gives r2  score to cross verify accuracy of model
```

```python
#checking the r2_score to y_test,y_predict to predict the accuracy
r_square=metrics.r2_score(y_test,y_pred)
print('R-square error for decision tree regessor is:',round(100*(r_square),3))
```

# Rework on different models
Random forest regressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
rfr = RandomForestRegressor(random_state=5)
rfr.fit(x_train,y_train)
RandomForestRegressor(random_state=5)
```

```python
y1_pred = rfr.predict(x_test)
```

```python
mae = metrics.mean_absolute_error(y_test, y1_pred)
print('Mean Absolute Error (MAE):', mae)

# Mean Squared Error (MSE)
mse = metrics.mean_squared_error(y_test, y1_pred)
print('Mean Squared Error (MSE):', mse)

# Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)
print('Root Mean Squared Error (RMSE):', rmse)

# Mean Absolute Percentage Error (MAPE)
mape = np.mean(np.abs((y_test - y1_pred) / y_test)) * 100
print('Mean Absolute Percentage Error (MAPE):', round(mape, 2))

# Accuracy
accuracy = 100 - mape
print('Accuracy:', round(accuracy, 2))
```

# Cross Validation
cross validation is a resampling method that uses different position of the data to test and learn a model on different direction

```python
from sklearn.model_selection import ShuffleSplit
model = LinearRegression()
ssplit=ShuffleSplit(n_splits=10,test_size=0.30)
from sklearn.model_selection import cross_val_score
results=cross_val_score(model,x,y,cv=ssplit)
print(results)
print(np.mean(results))
```

as we can see that for different iteration we are getting accuracy of almost 100%
